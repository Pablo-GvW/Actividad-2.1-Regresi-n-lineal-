```python
def limpiar_porcentaje(x):
    import pandas as pd
    if isinstance(x, str) and '%' in x:
        return float(x.strip('%')) / 100
    return pd.to_numeric(x, errors='coerce')


df['host_acceptance_rate'] = df['host_acceptance_rate'].apply(limpiar_porcentaje)
df['host_response_rate'] = df['host_response_rate'].apply(limpiar_porcentaje)


df['price'] = df['price'].replace('[$,]', '', regex=True).astype(float)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('listings copia.csv')

df['host_acceptance_rate'] = df['host_acceptance_rate'].replace('%', '', regex=True).astype(float) / 100
df['host_response_rate'] = df['host_response_rate'].replace('%', '', regex=True).astype(float) / 100
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['host_is_superhost'] = df['host_is_superhost'].map({'t': 1, 'f': 0})
df['room_type'] = df['room_type'].astype('category').cat.codes


columnas_modelo = [
    'host_id', 'host_acceptance_rate', 'host_is_superhost',
    'host_total_listings_count', 'room_type', 'accommodates',
    'bedrooms', 'price', 'review_scores_value', 'reviews_per_month'
]

df_modelo = df[columnas_modelo].dropna()
correlacion = df_modelo.corr()

```

```python
# host_id como variable dependiente
df1 = df_modelo.dropna(subset=columnas_modelo)
X1 = df1.drop(columns=['host_id'])
y1 = df1['host_id']
modelo1 = LinearRegression().fit(X1, y1)
r2_1 = r2_score(y1, modelo1.predict(X1))
coef_1 = pd.Series(modelo1.coef_, index=X1.columns)
sns.heatmap(coef_1.to_frame().T, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor - host_id')
plt.show()

# host_acceptance_rate como variable dependiente
X2 = df1.drop(columns=['host_acceptance_rate'])
y2 = df1['host_acceptance_rate']
modelo2 = LinearRegression().fit(X2, y2)
r2_2 = r2_score(y2, modelo2.predict(X2))
coef_2 = pd.Series(modelo2.coef_, index=X2.columns)
sns.heatmap(coef_2.to_frame().T, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('host_acceptance_rate')
plt.show()

# host_is_superhost como variable dependiente
X3 = df1.drop(columns=['host_is_superhost'])
y3 = df1['host_is_superhost']
modelo3 = LinearRegression().fit(X3, y3)
r2_3 = r2_score(y3, modelo3.predict(X3))
coef_3 = pd.Series(modelo3.coef_, index=X3.columns)
sns.heatmap(coef_3.to_frame().T, annot=True, cmap='Blues', fmt=".2f")
plt.title('host_is_superhost')
plt.show()



# host_total_listings_count como variable dependiente
X4 = df1.drop(columns=['host_total_listings_count'])
y4 = df1['host_total_listings_count']
modelo4 = LinearRegression().fit(X4, y4)
r2_4 = r2_score(y4, modelo4.predict(X4))
coef_4 = pd.Series(modelo4.coef_, index=X4.columns)
sns.heatmap(coef_4.to_frame().T, annot=True, cmap='Purples', fmt=".2f")
plt.title('host_total_listings_count')
plt.show()


# room_type como variable dependiente
X5 = df1.drop(columns=['room_type'])
y5 = df1['room_type']
modelo5 = LinearRegression().fit(X5, y5)
r2_5 = r2_score(y5, modelo5.predict(X5))
coef_5 = pd.Series(modelo5.coef_, index=X5.columns)
sns.heatmap(coef_5.to_frame().T, annot=True, cmap='Oranges', fmt=".2f")
plt.title('room_type')
plt.show()

# accommodates como variable dependiente
X6 = df1.drop(columns=['accommodates'])
y6 = df1['accommodates']
modelo6 = LinearRegression().fit(X6, y6)
r2_6 = r2_score(y6, modelo6.predict(X6))
coef_6 = pd.Series(modelo6.coef_, index=X6.columns)
sns.heatmap(coef_6.to_frame().T, annot=True, cmap='Reds', fmt=".2f")
plt.title('accommodates')
plt.show()

# bedrooms como variable dependiente
X7 = df1.drop(columns=['bedrooms'])
y7 = df1['bedrooms']
modelo7 = LinearRegression().fit(X7, y7)
r2_7 = r2_score(y7, modelo7.predict(X7))
coef_7 = pd.Series(modelo7.coef_, index=X7.columns)
sns.heatmap(coef_7.to_frame().T, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('bedrooms')
plt.show()

# price como variable dependiente
X8 = df1.drop(columns=['price'])
y8 = df1['price']
modelo8 = LinearRegression().fit(X8, y8)
r2_8 = r2_score(y8, modelo8.predict(X8))
coef_8 = pd.Series(modelo8.coef_, index=X8.columns)
sns.heatmap(coef_8.to_frame().T, annot=True, cmap='BuGn', fmt=".2f")
plt.title('price')
plt.show()

# review_scores_value como variable dependiente
X9 = df1.drop(columns=['review_scores_value'])
y9 = df1['review_scores_value']
modelo9 = LinearRegression().fit(X9, y9)
r2_9 = r2_score(y9, modelo9.predict(X9))
coef_9 = pd.Series(modelo9.coef_, index=X9.columns)
sns.heatmap(coef_9.to_frame().T, annot=True, cmap='YlOrRd', fmt=".2f")
plt.title('review_scores_value')
plt.show()

# reviews_per_month como variable dependiente
X10 = df1.drop(columns=['reviews_per_month'])
y10 = df1['reviews_per_month']
modelo10 = LinearRegression().fit(X10, y10)
r2_10 = r2_score(y10, modelo10.predict(X10))
coef_10 = pd.Series(modelo10.coef_, index=X10.columns)
sns.heatmap(coef_10.to_frame().T, annot=True, cmap='Blues', fmt=".2f")
plt.title('review_scores_value')
plt.show()


print(" host_id")
print("R²:", round(r2_1, 4))
print("Coeficientes:")
print(coef_1)

print("host_acceptance_rate")
print("R²:", round(r2_2, 4))
print("Coeficientes:")
print(coef_2)

print("host_is_superhost")
print("R²:", round(r2_3, 4))
print("Coeficientes:")
print(coef_3)

print("host_is_superhost")
print("R²:", round(r2_4, 4))
print("Coeficientes:")
print(coef_4)

print("host_is_superhost")
print("R²:", round(r2_5, 4))
print("Coeficientes:")
print(coef_5)

print("host_is_superhost")
print("R²:", round(r2_6, 4))
print("Coeficientes:")
print(coef_6)

print("host_is_superhost")
print("R²:", round(r2_7, 4))
print("Coeficientes:")
print(coef_7)

print("host_is_superhost")
print("R²:", round(r2_8, 4))
print("Coeficientes:")
print(coef_8)

print("host_is_superhost")
print("R²:", round(r2_9, 4))
print("Coeficientes:")
print(coef_9)

print("host_is_superhost")
print("R²:", round(r2_10, 4))
print("Coeficientes:")
print(coef_10)
```
